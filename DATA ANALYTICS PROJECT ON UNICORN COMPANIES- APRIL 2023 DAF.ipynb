## Data Cleaning, Manipulation and Transformation

```python
# Converting the Valuation column to integer for Data manipulation

data['Valuation'] = data['Valuation'].astype(str).str.replace('$', '', regex=False).str.replace('B', '', regex=False)
```

```python
data
```

```python
data['Valuation'] = pd.to_numeric(data['Valuation'], errors='coerce').astype(int)
```

```python
# Converting the Funding column to integer in Billions for Data manipulation and creating a new column for the new values

data['Funding'] = data['Funding'].astype(str).str.replace('$', ' ', regex=False).str.replace('Unknown', ' ', regex=False)

funding_new_list = []
for val in data['Funding']:
    if val.endswith('B'):
        funding_new_list.append(int(val[:-1]) * 1)
    elif val.endswith('M'):
        funding_new_list.append(int(val[:-1]) * 0.001)
    else:
        funding_new_list.append(None)
data['Funding_new'] = funding_new_list
```

```python
data
```

```python
# Creating a new column measuring the Return on Investment

data['Return on Investment'] = data['Valuation'] - data['Funding_new']
data
```

```python
# Creating a new column showing the Year of date joined, for easy data manipulation

data['Year_Date_Joined'] = data['Date Joined'].str.slice(stop = 4)
data
```

```python
# Converting the 'Year Date Joined' column to Integer type

data['Year_Date_Joined'] = pd.to_numeric(data['Year_Date_Joined'], errors='coerce').astype(int)
```

```python
# Creating a new column that shows how long it took each company to atain the Unicorn status

data['Period_To_Joining'] = data['Year_Date_Joined'] - data['Year Founded']
data
```

```python
# Calculate Age of each company and create a new column for same

current_year = 2023
data['Age'] = current_year - data['Year Founded']
```

```python
#Industries having highest ROI

# Filter for companies with positive ROI
data_positive_ROI = data[data['Return on Investment'] > 0]

# Group by industry and calculate median ROI
industry_ROI = data_positive_ROI.groupby('Industry')['Return on Investment'].max()

# Sort by median ROI in descending order
industry_ROI_sorted = industry_ROI.sort_values(ascending=False)

# Display the result
industry_ROI_sorted
```

### The Artificial Intelligence industry has the highest Return on Investment.

```python
#Correlation between Age of Companies, Fundings received and ROI

# Calculate age of each company
current_year = 2023
data['Age'] = current_year - data['Year Founded']

# Calculate correlation between age and ROI
correlation = data[['Age', 'Valuation', 'Funding_new', 'Return on Investment']].corr()

# Display the correlation matrix
correlation
```

```python
# Correlation matrix and heatmap

fig, ax = plt.subplots(figsize=(10,10))   
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.1, ax=ax)
plt.show()
```

### The Valuation of a company has a perfectly positive (strong) relationship with the Return on Investment of the company. Fundings received also showed a strong relationship with Return on Investment of Unicorns, as well as on the Valuation of the Unicorn.
